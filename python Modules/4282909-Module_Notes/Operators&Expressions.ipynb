## **Practice problems**
1. An aircraft tracking station determines the distance from a common point O to each aircraft and the angle between the aircrafts. If angle between the two aircrafts is equal to 49 degrees and the distances from point O to the two aircrafts are 50 km and 72 km, find distance d between the two aircrafts.(round answers to 1 decimal place).
(Hint: Apply cosine rule)

2. At t = pi/4, fine the value of y


Capture.PNG

```python
b = 50
c = 72

angle_degrees = 49

angle_radians = math.radians(angle_degrees)

a_square = (b ** 2) + (c ** 2) - (2 * b *c * math.cos(angle_radians))

a = math.sqrt(a_square)

print(round(a, 1))
```
