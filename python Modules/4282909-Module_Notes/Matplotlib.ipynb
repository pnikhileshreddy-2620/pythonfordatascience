# Introduction
* Matplotlib is a popular Python library used for creating high-quality static, animated, and interactive visualizations in 2D and 3D. It provides a wide range of functions and tools for visualizing data, and it is widely used in the scientific community and in data science for exploratory data analysis, data visualization, and presentation of results.

* Matplotlib can be used to create a variety of charts, plots, and graphs, including line charts, scatter plots, histograms, bar charts, and heatmaps, among others. It provides a lot of customization options, allowing users to control various aspects of their visualizations such as color, font, labels, and axes.

* The library is built on top of NumPy, a popular library for numerical computing in Python, and integrates well with other libraries like Pandas for data manipulation and SciPy for scientific computing. It also supports different output formats, including PNG, PDF, SVG, and interactive visualizations in Jupyter notebooks and web applications.

* Pyplot provides a range of functions to create different types of plots, such as line plots, scatter plots, bar plots, and histograms, among others. It also provides functions for labeling the axes, setting the title, adding legends, and adjusting the size and layout of the plot.

* The primary benefit of using Pyplot is that it makes it easy to create high-quality visualizations quickly, without needing to write a lot of code. This can save time and effort, particularly for data exploration and analysis, where it is important to quickly generate plots and iterate through different visualizations to gain insights into the data.

### Line plot

```python
fig, ax = plt.subplots()
Month = ["Jan", "Feb", "March"]
Av_temp = [10, 13, 27]

ax.plot(Month, Av_temp)
```

In the first example, fig, ax = plt.subplots() is used to create a figure object fig and an axes object ax. The plot method is then called on the ax object to create the line plot. This approach allows for more customization and control over the plot, such as creating multiple subplots within the figure.

```python
Month = ["Jan", "Feb", "March"]
Av_temp = [10, 13, 27]

plt.plot(Month, Av_temp)
```

In the second example, plt.plot(x, y) is used to create a line plot directly on the current axes. This approach is simpler and may be more convenient for creating quick plots or for cases where more customization is not needed.

## Barplot

```python
dt = {"Apple": 40, "Banana": 25, "Mango": 80, "Grapes": 45, "Pineapple": 25}
label = list(dt.keys())
values = list(dt.values())

plt.bar(label, height = values, edgecolor = 'c', facecolor = '#F0F8FF')
plt.xlabel("Fruits")
plt.ylabel("Price")
plt.title("Fruits prices")
plt.show()
```

```python
dt = {"Apple": 40, "Banana": 25, "Mango": 80, "Grapes": 45, "Pineapple": 25}
categories = list(dt.keys())
values = list(dt.values())

plt.barh(categories, width = values, edgecolor = 'c', facecolor = '#F0F8FF')
plt.xlabel("Fruits")
plt.ylabel("Price")
plt.title("Fruits prices")
plt.show()
```

## Pie chart

```python
covid_cases = {"AP": 1000, "TS": 1500, "UP": 3000, "MH": 5000}

# Get the labels and values from the dictionary
labels = list(covid_cases.keys())
values = list(covid_cases.values())

# Create the pie chart
plt.pie(values, labels=labels, autopct='%1.1f%%')

# Add a title
plt.title('COVID-19 Cases by State')

# Show the chart
plt.show()
```

```python
covid_cases = {"AP": 1000, "TS": 1500, "UP": 3000, "MH": 5000}

# Get the labels and values from the dictionary
labels = list(covid_cases.keys())
values = list(covid_cases.values())
exp = [0, 0.2, 0, 0]

# Create the pie chart
plt.pie(values, labels=labels, autopct='%1.1f%%',  explode = exp)

# Add a title
plt.title('COVID-19 Cases by State')

# Add a legend
plt.legend(loc = 3)

# Show the chart
plt.show()
```

## Histograms

```python
# Importing numpy
import numpy as np

# Creating random values between 19 and 60. total 100 value
ages = np.random.normal(loc=39.5, scale=7, size=100) # here mean is 39.5 and std dev is 7

plt.hist(ages)
plt.show()
```

```python
# Create histogram
plt.hist(ages, bins=15, color='skyblue', edgecolor='black')

# Add labels and title
plt.xlabel('Age in years')
plt.ylabel('Frequency')
plt.title('Histogram of Random Data')

# Show the plot
plt.show()
```

## Boxplot

```python
plt.boxplot(ages)
 
# Add labels and title
plt.xlabel('Age')
plt.ylabel('Age in years')
plt.title('Box plot of age')

# Show the plot
plt.show()
```

```python
data1 = np.random.normal(100, 20, 20)
data2 = np.random.normal(100, 40, 20)
data3 = np.random.normal(100, 60, 20)
 
# Creating plot
plt.boxplot([data1, data2, data3])
 
# show plot
plt.show()
```

## Subplots


```python
import numpy as np

# Create some data
x = np.linspace(0, 2*np.pi, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)

# Create a figure with 3 subplots
fig, axs = plt.subplots(3, 1) # 3 rows one column

# Plot the data on each subplot
axs[0].plot(x, y1)
axs[0].set_title('Sin(x)')
axs[1].plot(x, y2)
axs[1].set_title('Cos(x)')
axs[2].plot(x, y3)
axs[2].set_title('Tan(x)')

# Set the spacing between subplots
plt.subplots_adjust(hspace=0.5)

# Show the plot
plt.show()
```
