## High order function
A higher-order function in Python is a function that takes one or more functions as arguments or returns a function as its result.

## Map
This function takes a function and an iterable as arguments, and applies the function to each element in the iterable, returning a new iterable with the results.

```python
# define a list of numbers
numbers = [1, 2, 3, 4, 5]

# define a function to double a number
def double_number(num):
    return num * 2

# use map() to apply the function to each element in the list
doubled_numbers = list(map(double_number, numbers))

# print the result
print(doubled_numbers)
```

### reduce()

- Useful when you need to apply a function to an iterable and reduce it to a single cumulative value.



```python
# importing functools for reduce()
from functools import reduce
```

```python
def add(x, y):
    return x + y

numbers = [1, 2, 3, 4, 5]
total = reduce(add, numbers)
print(total) 
```
