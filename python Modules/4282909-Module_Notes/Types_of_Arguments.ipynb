# Types of Arguments/Parameters
1. Required arguments
2. Keyword arguments
3. Default arguments

## Required and positional arguments
* Number of arguments should be same in definition and call.
* Order should also be same

## Keyword arguments
* Number of arguments should be same in definition and call.
* Keyword arguments order can be any thing.
* But should be put after required arguments in the function call.

## Default arguments
* These are optional arguments.
* Should be put after the non-default arguments in the function definition.
* If not passed in function call, the default behaviour is followed.

```python
def demo(a, b, c):
  print(a)
  print(b)
```

```python
demo(3, 2, 4, 4)
```

```python
def demo(a, b, c):
  print("a :", a)
  print("b: ", b)
  print("c: ", c)
```

```python
demo(3, 2, 4)
```

```python
demo(c = 3, a = 4, b = 1)
```

```python
demo(3, c = 2, b = 10)
```

```python
def demo(a, b = 100):
  print("a: ", a)
  print("b: ", b)
```

```python
demo(3, 200)
```

```python
def area(length, width = None):
  if width == None:
    print("The area of square: ", length * length)
  else:
    print("The area of rectangle: ", length * width)
```

```python
area(5)
```
