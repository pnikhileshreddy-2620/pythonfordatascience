### 1. Simple **if**

```
Syntax :-   if (condition):
                Statement 1

                    .
                    .
                Statement n
            Statement x
```



````python
a = -3

if a > 0:
  print("Positive")
  print("Still in if block")
  
print("Out of if condition")
````

### 2. **if** - **else**

keywords - **if** & **else**

Two Blocks of code -> indented

**if** block = True block

**else** block = False block

```
Syntax :-   if (condition):
                Statement 1 
                Statement 2
                    .
                    .
                Statement n
            else:
                Statement 1 
                Statement 2
                    .
                    .
                Statement n
            Statement x
```



````python
a = 3

if a % 2 == 0:
  print("Its an even number")
else:
  print("Its an odd number")
````

### 3. Nested **if** - **else**

keywords - **if** & **else**

Multiple Blocks of code -> indented

**if** block = True block

**else** block = False block

Used when **multiple dependent conditions** are present

```
Syntax :-   if (condition 1):
                # True Block
                if (condition 2):
                   # Inner True block
                   Statement 1 
                   Statement 2
                else:
                   # Inner False block
                   Statement 1 
                   Statement 2
            else:
                # False block
                if (condition 3):
                   # Inner True block
                   Statement 1 
                   Statement 2
                else:
                   # Inner False block
                   Statement 1 
                   Statement 2
            Statement x
```

````python
# Nested if-else
# Find the largest of three numbers
a = int(input("Enter the first number: "))
b = int(input("Enter the second number: "))
c = int(input("Enter the third number: "))

# a = 10, b = 5, c = 20
# a = 5, b = 20, c =100
if (a > b):
  if (a > c):
    print("Largest number is: ", a)
  else:
    print("Largest number is: ", c)
else:
  if (b > c):
    print("Largest number is: ", b)
  else:
    print("Largest number is: ", c)
print("Immediate sequence statement")
````

### 4. **if** - **elif** - **else**

keywords - **if** , **elif**, **else**

Multiple Blocks of code -> indented

**if** block = True block

**elif** block = True block

**else** block = False block

Used when **multiple independent conditions** are present 

```
Syntax :-   if (condition 1):
                Statement 1 
                Statement 2
            elif (condition 2):
                Statement 1 
                Statement 2
                    .
                    .
            else:
                Statement 1
                Statement 2
            Statement x
```

````python
# if - elif - else
# Find the grade of the student
marks = int(input("Enter the marks of the student: "))
if marks > 90:
  print("Grade: A")
elif marks > 75: # 75 - 90
  print("Grade: B")
elif marks > 60: # 60 -75
  print("Grade: C")
elif marks > 45:
  print("Grade: D")
else:
  print("Grade: F")
````
