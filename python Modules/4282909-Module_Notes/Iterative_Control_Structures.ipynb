#  **Iterative control structures**

* Also called as Loops
* Iterative = repetition

Repeated execution of a block of statements = Loops

Number of repetitions = number of iterations

## Loop Control

1.   Loop variable initialization
2.   Condition for termination
3.   Updation of Loop variable


## Two types of Loops


1.   While Loop
2.   For Loop


# While Loop



```
Syntax: While (condition):
            Statement 1
            Statement 2
                .
                .
            Statement n
        Statement x
```

        

````python
# Loop variable initialization
# Condition for termination
# Updation of Loop variable
````

````python
# 5! = 5*4*3*2*1 = 120
````

````python
# Results initialization
# Product -> 1
# Sum -> 0
````

````python
# Example: Factorial
# Loop variable initialization
number = 5
fact = 1
while number > 0: # -> Condition for termination
  fact *= number # 120 * 1 = 120
  print("Result: ", fact, "number: ", number)
  number -= 1 # 0 -> Loop variable updation
  

print(fact)
````

# For Loop

```
Syntax: for i in sequence:
            Statement 1
            Statement 2
                .
                .
            Statement n
        Statement x
```
* All the 3 Loop controls (Initialization, termination and updation of Loop variable) are performed in the sequence.

* Sequence could be any collection (Array, List, tuple, dictionary).

````python
for loop_variable in "python":
  print(loop_variable)
````

````python
for i in [0, 1, 2, 3, 4, 5]: # -> List
  print(i)
````

````python
for i in range(1, 6):
  print(i)
````

````python
# 5! = 1*2*3*4*5
````

````python
# Example: Factorial

fact = 1

for i in range(1, 6):
  fact = fact * i
  print("Iteration: ", i, "Result: ", fact)

print(fact)
````

## **While vs For loop**
1. **While**: Indefinite, **For:** Definite
2. **While**: To be used when the number of iterations are unknown.    **For**: To be used when the number of iterations are known.

## **Break and Continue**

````python
# Break

for i in range(10):
  print(i)
  if i == 5:
    break

for i in range(10):
  for j in range(20):
    for k in range(30):
      if k == 2:
        break
````

````python
# Continue

for i in range(10):
  if i == 4:
    continue
  print(i)
````

````python
for i in "Qwerty@254":
  if i in "0123456789":
    continue
  print(i)
````

````python
count = 0
number = 0

while True:
  if number % 2 == 0:
    print(number)
    count += 1
    
  number += 1

  if count == 50:
    break
````

````python
print(count)
````
