# Seaborn

Seaborn is a Python data visualization library built on top of Matplotlib. It provides a high-level interface for creating informative and attractive statistical graphics.

Here are some basic features of Seaborn library:

1- **Distribution plots**: Seaborn provides a number of distribution plots, such as histogram, kernel density estimate (KDE) plot, rug plot, etc., to visualize the distribution of data.

2- **Scatter plots**: Seaborn provides several types of scatter plots, including scatter plot, hexbin plot, and regression plot, to visualize the relationship between two variables.

3- **Categorical plots**: Seaborn provides various types of categorical plots, such as box plot, violin plot, swarm plot, bar plot, etc., to visualize categorical data.

4- **Heatmaps**: Seaborn provides a heatmap function to create a color-coded matrix to visualize the correlation between variables.

5- **Pair plots**: Seaborn provides a pairplot function to create a matrix of scatter plots and histograms to visualize the pairwise relationships between variables.

6- **Style customization**: Seaborn allows customization of plot aesthetics by changing the default settings, such as color palette, font size, line style, etc.

Overall, Seaborn is a powerful tool for creating informative and visually appealing data visualizations in Python.







```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Dist plot

```python
sns.set(style="white") # setting background as white

d = np.random.normal(100, 20, 100)
  
sns.distplot(d, kde=True)
```

## Scatter plot

```python
tip = sns.load_dataset('tips') # laoding tips dataset
ax = sns.scatterplot(x='tip', y='total_bill', data=tip, marker = 'H', edgecolor='c', facecolor = 'pink')
ax.set(xlabel='tips', ylabel='total bill', title = 'Relationship between tips and total bill')
plt.show()
```

```python
tip = sns.load_dataset('tips')
sns.scatterplot(x='tip', y='total_bill', data=tip, marker = 'H', edgecolor='r', facecolor = 'pink')
plt.title("Relationship between tips and total bill")
plt.xlabel("tips")
plt.ylabel('Total bill')
```

```python
tip = sns.load_dataset('tips')
sns.scatterplot(x='tip', y='total_bill', hue = 'sex', data=tip, marker = 'H', edgecolor='r', facecolor = 'pink')
plt.title("Relationship between tips and total bill")
plt.xlabel("tips")
plt.ylabel('Total bill')
```

## Regression

```python
sns.set(font_scale=1.5, style="ticks")
  
ax = sns.lmplot(x='tip', y='total_bill', data=tip, fit_reg = True)
ax.set(xlabel='tips', ylabel='total bill', title = 'Relationship between tips and total bill')
plt.show()
  
```

## Line plot

```python
# Load example dataset (flights)
flights = sns.load_dataset('flights')

# Create line plot with markers, width, and color
sns.lineplot(data=flights, x='year', y='passengers', marker='o', markersize=8, linewidth=2, color='g', ci = None)

# Add labels and title
plt.xlabel('Year')
plt.ylabel('Passengers')
plt.title('Number of passengers over time')

# Show the plot
plt.show()
```

## Pair plot

```python
df = sns.load_dataset('tips')
sns.pairplot(df, hue ='sex')
plt.show()
```

## Heatmap

```python
# Load tips dataset
tips = sns.load_dataset("tips")

# Create correlation matrix
corr_matrix = tips.corr()

# Create heatmap
sns.heatmap(corr_matrix, annot=True)

# Show the plot
plt.show()
```

## Facet grid

```python
sns.FacetGrid(tip, col="day", row="sex").map(sns.histplot, "total_bill")
```

```python
# Create a FacetGrid plot with different colors for male and female customers
g = sns.FacetGrid(tips, col="day", row="sex", hue="sex")
g.map(sns.histplot, "total_bill", alpha=0.7)
g.add_legend()

# Add labels and title
g.set_axis_labels("Total Bill", "Count")
g.fig.suptitle("Total Bill by Day and Sex", y=1.05)

# Show the plot
plt.show()
```
